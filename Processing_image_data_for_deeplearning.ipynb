# Resizing image using pillow library
* sometimes we maight have lots of image with different size so we need resizing images to make all images to the same image size

```python
from PIL import Image
```

```python
img= Image.open('/content/the_view.jpg')
img_resized=img.resize((200,200))
```

```python
img_resized.save('the_veiw_resized.jpg')
```

```python
#displaying the image from numpy array
img_res=mpimg.imread('/content/the_veiw_resized.jpg')
img_res_plot=plt.imshow(img_res)
plt.show()
```

#**How to convert the RGB image to grayscale image**
* openCV 

```python
#importing openCV library
import cv2
```

```python
img=cv2.imread('/content/the_view.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale)
```

```python
grayscale.shape
```

cv2.imshow() will display the image but this is not allowed on Google colab.

```python
#saving the grayscale image
cv2.imwrite('the_view_grayscale.jpg',grayscale)
```
